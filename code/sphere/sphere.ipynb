## The Sphere $S^2$

```python
# compute Fisher information matrix and metric
import sympy as sp
from sympy import Function, diff, simplify, integrate, preview, Matrix, lambdify, Symbol
from sympy import symbols, sin, cos, pi, Rational, Integer
from sympy.diffgeom import Manifold, Patch, CoordSystem, TensorProduct
from sympy.diffgeom.rn import * # we'll use the r2_* and r3_*
TP = TensorProduct

# manifold and coord systems
M = R2
dim = M.dim
patch = Patch('P',M)

coords_r = R2_r.coord_functions() #[Dummy(s) for s in ['x', 'y']]
x = coords_r[0]; y = coords_r[1]

R2_sphere = CoordSystem('sphere',patch,['theta','pi'])
coords_s = R2_sphere.coord_functions()
theta = coords_s[0]; phi = coords_s[1]
dtheta, dphi = [Dummy(s) for s in ['theta', 'phi']]
R2_sphere.connect_to(R2_r,[dtheta,dphi],[sin(dtheta)*cos(dphi),sin(dtheta)*sin(dphi)])
J = R2_sphere.jacobian(R2_r,coords_s) # Jacobian matrix of transformation

# map F: R2_sphere\rightarrow R3_s
radius = 1
Fs = Matrix([radius,theta,phi])
Fr = Fs.subs(zip(coords_s,R2_sphere.coord_tuple_transform_to(R2_r,coords_r)))
FR3r = Matrix([sin(theta)*cos(phi),sin(theta)*sin(phi),cos(theta)]).subs(zip(coords_s,coords_r))
JFr = Matrix(3,2,lambda i,j: R2_r.base_vector(j)(Fr[i]))

# map F stereographic R2_r\rightarrow R3_s
Fst = Matrix([2*x/(1+x**2+y**2),2*y/(1+x**2+y**2),-(-1+x**2+y**2)/(1+x**2+y**2)])
FR3st = Fst
JFst = Matrix(3,2,lambda i,j: R2_r.base_vector(j)(Fst[i]))

# choose coords
#F = Fr; JF = JFr; FR3 = FR3r
F = Fst; JF = JFst; FR3 = FR3st
```

```python
# sympy/diffgeom stuff
# sympy messes up, let's try helping it out
dummyx = Symbol('x')
dummyy = Symbol('y')
dcoords = [dummyx,dummyy] # dummy coordinates
# below functions changes back and forth between diffgeom symbols and dcoords
todcoords = lambda S: simplify(S.subs(zip(coords_r,dcoords)))
fromdcoords = lambda S: S.subs(zip(dcoords,coords_r))
# lambdify
import numpy as np
toArray = lambda A: np.double(np.array(A))
lambd = lambda S : (lambda v: toArray(lambdify(coords_r,S)(*tuple(v))))

# find metric matrix and simplify
F = todcoords(F)
JF = todcoords(JF)

# metric matrix
GM = JF.T*JF
GM = simplify(GM)

# convert back
F = fromdcoords(F)
JF = fromdcoords(JF)
GM = fromdcoords(GM)

# evaluation
Fx = lambd(F)
JFx = lambd(JF)
GMx = lambd(GM)

F, JF, GM
```

```python
# metric tensor
dG = TP(0)
for i,j in itertools.product(range(dim),range(dim)):
    dG += GM[i,j]*TP(R2_r.base_oneforms()[i],R2_r.base_oneforms()[j])
dG
```

## Geodesics: Numerical Integration

```python
# compute geodesics
import numpy as np
from scipy import linalg
from scipy.integrate import odeint
from scipy.optimize import minimize, root, check_grad, approx_fprime
from functools import partial
import matplotlib.pyplot as plt
from sympy.diffgeom import metric_to_Christoffel_2nd, twoform_to_matrix
from sympy import lambdify

def intGeodesic(M, metric, p, v, Fr=np.empty([M.dim]), T=1., N_t=20):
    """ M ... """

    def geodesicf(M, C2ndf, x, t):
        """ input 2*M.dim vector x """ 
    
        gamma = x[0:M.dim] 
        dgamma = x[M.dim:2*M.dim]
        Fgamma = x[2*M.dim:].reshape([M.dim,-1])
    
        C2nd = C2ndf(gamma)
    
        d2gamma = -np.einsum('abc,b,c->a',C2nd,dgamma,dgamma) # geodesic equation
        dFgamma = -np.einsum('abc,b,cd->ad',C2nd,dgamma,Fgamma) # parallel transport
    
        return np.hstack( (dgamma,d2gamma, dFgamma.flatten()) )

    coord_sys = dG.atoms(CoordSystem).pop()
    
    # get Christoffel symbols
    C2nd = metric_to_Christoffel_2nd(metric)
    C2ndfun = lambd(C2nd)

    t_span = np.linspace(0., T, N_t) 
    gamma_t = odeint(partial(geodesicf, M, C2ndfun), np.hstack((p,v,Fr.flatten())), t_span)

    return (gamma_t[:,0:M.dim], gamma_t[:,M.dim:2*M.dim], gamma_t[:,2*M.dim:].reshape([-1,M.dim,M.dim])) # return points, tangent vectors, and Frame
```

```python
# compute mpps
import numpy as np
from scipy import linalg
from scipy.integrate import odeint
from scipy.optimize import minimize, root, check_grad, approx_fprime
from functools import partial
import matplotlib.pyplot as plt
from sympy.diffgeom import metric_to_Christoffel_2nd, twoform_to_matrix
from sympy import lambdify

def splitMomentum(xi):
    return (xi[0:dim],xi[dim:dim+dim**2].reshape([dim,dim]))
def concatMomentum(xi,xia):
    return np.hstack( (xi,xia.flatten()) )

# Hamiltonian
def H(x,Xa,xi,xia):
    delta = np.eye(dim)
    Gamma = lambd(C2nd)(x)
    GammaX = np.einsum('hja,ag->hgj',Gamma,Xa)
    W = np.einsum('ab,ka,lb->kl',delta,Xa,Xa)        
    dx = np.einsum('ij,j->i',W,xi)-np.einsum('ih,jbh,jb->i',W,GammaX,xia)
    dXa = -np.einsum('iah,hj,j->ia',GammaX,W,xi)+np.einsum('iak,kh,jbh,jb->ia',GammaX,W,GammaX,xia)
        
    return .5*(np.einsum('i,i->',dx,xi)+np.einsum('ia,ia->',dXa,xia))

def intMPP(M, metric, p, xi, Fr, T=1., N_t=20):
    """ M ... """
    
#    dGM = simplify(GM.subs([(coords_r[0],dx),(coords_r[1],dy)])) 
#    invdGM = dGM.inv()
#    DGM = [[[diff(dGM[i,j],dcoords[l]) for l in range(dim)] for j in range(dim)] for i in range(dim)]
#    DinvGM = [[[diff(invdGM[i,j],dcoords[l]) for l in range(dim)] for j in range(dim)] for i in range(dim)]
    
    def mppf(M, C2ndf, DC2ndf, xs, t):
        """ input 2*M.dim+2*M.dim**2 vector x """ 
    
        x = xs[0:M.dim] 
        Xa = xs[M.dim:M.dim+M.dim**2].reshape([M.dim,M.dim])
        xi = xs[M.dim+M.dim**2:2*M.dim+M.dim**2]
        xia= xs[2*M.dim+M.dim**2:2*M.dim+2*M.dim**2].reshape([M.dim,M.dim])
    
        C2nd = C2ndf(x)
        DC2nd = DC2ndf(x)
        Gamma = C2nd
        d0Gamma = DC2nd
        delta = np.eye(dim)
        
        #GMx =  toArray([[dGM[i,j].subs(zip(dcoords,x)) for j in range(dim)] for i in range(dim)])
        #invGMx =  toArray([[invdGM[i,j].subs(zip(dcoords,x)) for j in range(dim)] for i in range(dim)])
        #DGMx = toArray([[[DGM[i][j][l].subs(zip(dcoords,x)) for l in range(dim)] for j in range(dim)] for i in range(dim)])
        #DinvGMx = toArray([[[DinvGM[i][j][l].subs(zip(dcoords,x)) for l in range(dim)] for j in range(dim)] for i in range(dim)])

        # debug
        ## lagrangian
        #dx = xi
        #dXa = -np.einsum('abc,b,cd->ad',C2nd,dx,Xa) # parallel transport
        #dxi = -np.einsum('abc,b,c->a',C2nd,dx,dx) # geodesic equation
        #dxia = np.zeros([M.dim,M.dim])        
        
        ## Hamiltonian
        #dx = np.einsum('ij,j->i',invGMx,xi)
        #dXa = -np.einsum('abc,b,cd->ad',C2nd,dx,Xa) # parallel transport
        #dxi = -.5*np.einsum('bca,b,c->a',DinvGMx,xi,xi) # Hamiltonian version
        #dxia = np.zeros([M.dim,M.dim])

        # MPP
        GammaX = np.einsum('hja,ag->hgj',Gamma,Xa)
        d0GammaX = np.einsum('hjak,ag->hgjk',d0Gamma,Xa)        
        d1GammaX = np.einsum('xa,ihl->iahlx',delta,Gamma)
        
        W = np.einsum('ab,ka,lb->kl',delta,Xa,Xa)        
        d1W = np.einsum('il,jx->ijlx',delta,Xa)+np.einsum('jl,ix->ijlx',delta,Xa)
        
        d0g00 = np.zeros([M.dim,M.dim,M.dim])
        d0g01 = -np.einsum('ih,jbhl->ijbl',W,d0GammaX)
        d0g10 = -np.einsum('iahl,hj->iajl',d0GammaX,W)
        d0g11 = (np.einsum('iakl,kh,jbh->iajbl',d0GammaX,W,GammaX)+\
                 np.einsum('iak,kh,jbhl->iajbl',GammaX,W,d0GammaX))
        
        d1g00 = d1W
        d1g01 = -np.einsum('ihlx,jbh->ijblx',d1W,GammaX)-np.einsum('ih,jbhlx->ijblx',W,d1GammaX)
        d1g10 = -np.einsum('iahlx,hj->iajlx',d1GammaX,W)-np.einsum('iah,hjlx->iajlx',GammaX,d1W)
        d1g11 = (np.einsum('iaklx,kh,jbh->iajblx',d1GammaX,W,GammaX)+\
                 np.einsum('iak,khlx,jbh->iajblx',GammaX,d1W,GammaX)+\
                 np.einsum('iak,kh,jbhlx->iajblx',GammaX,W,d1GammaX))
        
        dx = np.einsum('ij,j->i',W,xi)-np.einsum('ih,jbh,jb->i',W,GammaX,xia)
        dXa = -np.einsum('iah,hj,j->ia',GammaX,W,xi)+np.einsum('iak,kh,jbh,jb->ia',GammaX,W,GammaX,xia)
        dxi = -.5*(np.einsum('hki,h,k->i',d0g00,xi,xi)+\
                   np.einsum('hkdi,h,kd->i',d0g01,xi,xia)+\
                   np.einsum('hgki,hg,k->i',d0g10,xia,xi)+\
                   np.einsum('hgkdi,hg,kd->i',d0g11,xia,xia))
        dxia = -.5*(np.einsum('hkia,h,k->ia',d1g00,xi,xi)+\
                    np.einsum('hkdia,h,kd->ia',d1g01,xi,xia)+\
                    np.einsum('hgkia,hg,k->ia',d1g10,xia,xi)+\
                    np.einsum('hgkdia,hg,kd->ia',d1g11,xia,xia))
        
        #print "energy: %s" % (np.einsum('i,i->',dx,xi)+np.einsum('ia,ia->',dXa,xia))
        
        return np.hstack( (dx,dXa.flatten(),dxi,dxia.flatten()) )

    coord_sys = dG.atoms(CoordSystem).pop()
    
    # get Christoffel symbols
    C2nd = metric_to_Christoffel_2nd(metric)
    # convert back and forth to take derivatives
    C2nd = [[[simplify(C2nd[i][j][k].subs(zip(coords_r,dcoords)))
               for k in range(M.dim)]
                for j in range(M.dim)]
                 for i in range(M.dim)]
    DC2nd = [[[[simplify(diff(C2nd[i][j][k],dcoords[l]))
               for l in range(M.dim)]
                for k in range(M.dim)]
                 for j in range(M.dim)]
                  for i in range(M.dim)]
    C2nd = [[[C2nd[i][j][k].subs([zip(dcoords,coords_r)])
               for k in range(M.dim)]
                for j in range(M.dim)]
                 for i in range(M.dim)]
    C2ndfun = lambd(C2nd)
    DC2ndfun = lambd(DC2nd)

    t_span = np.linspace(0., T, N_t) 
    x = p
    Xa = Fr
    if xi.size == M.dim:
        xia = np.zeros(Fr.shape)
    else:
        (xi,xia) = splitMomentum(xi)
    
    gamma_t = odeint(partial(mppf, M, C2ndfun, DC2ndfun), np.hstack((x,Xa.flatten(),xi,xia.flatten())), t_span)

    return (gamma_t[:,0:M.dim],\
            gamma_t[:,M.dim:M.dim+M.dim**2].reshape([-1,M.dim,M.dim]),\
            gamma_t[:,M.dim+M.dim**2:2*M.dim+M.dim**2],\
            gamma_t[:,2*M.dim+M.dim**2:2*M.dim+2*M.dim**2].reshape([-1,M.dim,M.dim])\
            ) # return points and frames
```

```python
import dill
from pathos.multiprocessing import ProcessingPool
from pathos.multiprocessing import cpu_count
P = ProcessingPool(cpu_count()/2)

def DiffSource(M, metric, ps, tol=1e-4, maxIter=20, x0=None, Xa0=None):

    def MPPLogf((idx,m)):
        #print "computing Log for sample %d" % idx
                
        Nsamples = (m.shape[0]-(M.dim+M.dim**2))/(M.dim+M.dim**2)
        x0 = (1./Nsamples)*m[0:M.dim]
        Xa0 = (1./Nsamples)*m[M.dim:M.dim+M.dim**2]
        Logsamples = m[M.dim+M.dim**2:].reshape([-1,M.dim+M.dim**2])
        xi = Logsamples[idx,:]

        (x_t, _, _, _) = intMPP(M, metric, x0, xi, Xa0, T=1.)

        res = ps[idx,:]-x_t[-1,:]

        (xi,xia) = splitMomentum(xi)
        resE = np.einsum('i,i->',res,res)
        EH = H(p,Fr,xi,xia)
        
        #print "computed Log for sample %d" % idx
        
        return weightH*EH+weightRes*resE
    
    
    def f(m):
        # parallel compute distances        
        Nsamples = (m.shape[0]-(M.dim+M.dim**2))/(M.dim+M.dim**2)

        input_args = zip(*(xrange(Nsamples), itertools.cycle((m,))))
        sol = P.imap(MPPLogf, input_args)
        Logs = np.array(list(sol))            
        #Logs = np.empty(Nsamples)
        #for i in range(Nsamples):
        #   Logs[i] = MPPLogf(input_args[i])

        res = (1./Nsamples)*np.sum(Logs)
        x0 = (1./Nsamples)*m[0:M.dim]
        Xa0 = (1./Nsamples)*m[M.dim:M.dim+M.dim**2].reshape([M.dim,M.dim])
        Gx0M = toArray(GM.subs(zip(coords_r,x0)));
        Xa02inner = np.einsum('ba,bi,ij->aj',Xa0,Gx0M,Xa0)
        detXa02 = np.linalg.det(Xa02inner)
        #print "f x0: %s, Xa02 diag: %s, Xa02 off: %s, det: %g, res %g" % (x0,np.diag(Xa02inner),Xa02inner[0,1],detXa02,res,)
        return res
    
    def constr(m):
        # constraints on frame
        Nsamples = (m.shape[0]-(M.dim+M.dim**2))/(M.dim+M.dim**2)
        x0 = (1./Nsamples)*m[0:M.dim]
        Xa0 = (1./Nsamples)*m[M.dim:M.dim+M.dim**2].reshape([M.dim,M.dim])
        Gx0M = toArray(GM.subs(zip(coords_r,x0)));
        Xa02inner = np.einsum('ba,bi,ij->aj',Xa0,Gx0M,Xa0)
        detXa02 = np.linalg.det(Xa02inner)
        
        res = -np.sum(np.abs([Xa02inner[0,1],Xa02inner[1,0],1-detXa02]))
        #print "constr res: %s" % res
        
        return res

    # number of samples
    Nsamples = ps.shape[0]
            
    if x0 == None:
        # initial point
        x0 = np.mean(ps,0)
    if Xa0 == None:
        # initial frame        
        Gx0M = toArray(GM.subs(zip(coords_r,x0)));
        Xa0 = np.eye(dim)
        Xa0 = np.linalg.solve(linalg.sqrtm(Gx0M),Xa0) # make orthonormal
    else:
        # make frame determinant one
        Xa02inner = np.einsum('ba,bi,ij->aj',Xa0,GpM,Xa0)
        detXa02inner = np.linalg.det(Xa02inner)
        print "determinant pre normalization: %g" % detXa02inner
        nf=detXa02inner**(-1./dim**2)
        Xa0 = nf*Xa0
        Xa02inner = np.einsum('ba,bi,ij->aj',Xa0,GpM,Xa0)
        detXa02inner = np.linalg.det(Xa02inner)
        print "determinant post normalization: %g" % detXa02inner

    print "initial point/frame, x0: %s, Xa0: %s" % (x0,Xa0)

    initval = np.hstack( (Nsamples*x0,Nsamples*Xa0.flatten(),np.zeros((Nsamples,M.dim+M.dim**2)).flatten(),) )
    res = minimize(f, initval, method='COBYLA',\
                   tol=tol,\
                   constraints={'type': 'ineq', 'fun': constr},\
                   options={'disp': True, 'maxiter': maxIter},\
                   )
    
    if not res.success:
        print "mean/covar optimization failed:\n%s" % res
    mu = (1./Nsamples)*res.x[0:M.dim]
    SigmaSQRT = (1./Nsamples)*res.x[M.dim:M.dim+M.dim**2]
    Logyis = res.x[M.dim+M.dim**2:].reshape([Nsamples,M.dim+M.dim**2])

    return (mu,SigmaSQRT,Logyis)
```
